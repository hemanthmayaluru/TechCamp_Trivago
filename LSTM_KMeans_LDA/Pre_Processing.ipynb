```python
import numpy as np
import pandas as pd
```

```python
'''Loading the input and pre-processing'''
df_new = pd.read_csv('Final sample Gettech.csv',sep=';',index_col=0)
df_test = df_new.groupby(['accommodation_id','basename','at','description','value_type_id'])['amenities_id'].apply(list)
df_test = df_test.to_frame().reset_index()
df_test2 = df_new.groupby(['accommodation_id','basename','at','description','value_type_id'])['amenities_cont'].apply(list)
df_test2 = df_test2.to_frame().reset_index()
final_df = pd.merge(df_test, df_test2[["accommodation_id", "amenities_cont"]], on="accommodation_id", how="left")
final_df.to_csv('processed_csv_file.csv', sep='\t', encoding='utf-8')
final_df.head()
```

```python
type_1_list = [6,8,50]
type_0_list = [2,3,4,5,9,11,49,53,54]
for elem in type_1_list:
    final_df.loc[final_df['at'] == elem, 'label'] = 1
for elem in type_0_list:
    final_df.loc[final_df['at'] == elem, 'label'] = 0
final_df.head()
```

```python
sorted_list = []
element_list = [13,17,46,47,48,63,255]
for li in final_df["amenities_id"]:
    new_li = []
    for elem in element_list:
        if elem in li:
            new_li.append(elem)
        else:
            new_li.append(0)
    sorted_list.append(new_li)
final_df["sorted_amenities_id"] = sorted_list
final_df.head()
```

```python
final_df[['a1','a2','a3','a4','a5','a6','a7']] = pd.DataFrame(final_df.sorted_amenities_id.values.tolist(), index= final_df.index)
#final_df.to_csv("final_pre-processed_input.csv", sep=';', encoding='utf-8')
final_df.head(20)
```
